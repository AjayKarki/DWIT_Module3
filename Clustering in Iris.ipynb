```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris['DESCR']
```

```python
x = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
x.head()
```

```python
y=  pd.DataFrame(iris.target,columns = ['target_names'])
```

```python
y.head()
```

```python
plt.figure(figsize=(12,3))
colors =  np.array(['red','green','blue'])
plt.subplot(2,2,1)
plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=colors[y['target_names']],s=40)
plt.title('SepalLengthVsSepalWidth')
plt.xlabel('Sepal Len')
plt.subplot(2,2,2)
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=colors[y['target_names']],s=40)
plt.title('PetalLengthVsPetalWidth')
plt.xlabel('Petal Len')
plt.subplot(2,2,3)
plt.scatter(x['sepal length (cm)'],x['petal width (cm)'],c=colors[y['target_names']],s=40)
plt.title('SepalLengthVsSepalWidth')
plt.xlabel('Sepal Len')
plt.subplot(2,2,4)
plt.scatter(x['sepal length (cm)'],x['petal length (cm)'],c=colors[y['target_names']],s=40)
plt.title('PetalLengthVsPetalWidth')
plt.xlabel('Petal Len')
```

```python
model = KMeans(n_clusters=3)
x= x[['petal length (cm)','petal width (cm)']]
model.fit(x)
```

```python
model.labels_
```

```python
plt.clf()
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
colors =  np.array(['red','green','blue'])

plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=colors[y['target_names']],s=40)
plt.title('Before CLassification')

plt.subplot(1,2,2)
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=colors[model.labels_],s=40)
plt.title('After CLassification')
```
