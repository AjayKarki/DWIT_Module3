```python
x= iris.data
y=iris.target
```

```python
np.cov(x.T)
```

```python
mean = x.mean(axis=0)
```

```python
mean
```

```python
x_std = x-mean
```

```python
x_std
```

```python
cov = np.cov(x_std.T)
```

```python
cov
```

```python
en_v,en_vect = np.linalg.eig(cov)
```

```python
en_v
```

```python
tm = en_vect[:,:3]
```

```python
tm
```

```python
x_trans = np.dot(x_std,tm)
```

```python
x_trans
```

```python
x_re = np.dot(x_trans,tm.T)+mean
```

```python
x_re
```

```python
error = ((x-x_re)**2).mean()
```
