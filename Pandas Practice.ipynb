### Inplace
Chhanges dataframe without need of assignment

```python
df =  pd.DataFrame(np.eye(3),columns =  ['a','b','c'])
df_2 = pd.DataFrame({
    'a':[1],
    'b':[1],
    'c':[1]
})
print(df)
print(df_2)
df = df.append(df_2)
df.reset_index(inplace=True,drop=True)
print(df)
```

```python
df.query('b==0')
```

```python
df.query('b in [0,1]')
```

```python
q=[0,1]
df.query('b in @q')
```

```python
df[3,'a']=90
df
```

```python
df = pd.DataFrame([[1,2],[3,4]],columns = ['a','b'])
df2 = pd.DataFrame([[5,6],[7,8]],columns = ['a','b'])
df = df.append(df2).reset_index(drop=True)
df = df.drop(0)
df
```
