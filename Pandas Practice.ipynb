## Joins

```python
import pandas as pd
left =  pd.DataFrame({
    'id':[1,2,3,4],
    'Name':['Shreyansh','Ram','Shyam','Hari'],
    'subject':['Nepali','Nepali','Math',"Computer"]
})
right = pd.DataFrame({
    'id':[1,2,3,4,5],
    'Age': [51,22,33,44,55]
})
pd.merge(left,right,on='id',how='outer')
```

```python
left =  pd.DataFrame({
    'id':[1,2,3,4],
    'Name':['Shreyansh','Ram','Shyam','Hari'],
    'subject':['Nepali','Nepali','Math',"Computer"]
})
right = pd.DataFrame({
    'id':[1,2,3,4,5],
    'Age': [51,22,33,44,55],
    'subject':['Nepali','Nepali',"Computer",'Math','Science']

})
pd.merge(left,right,on=['id','subject'])
```

```python
pd.merge(left,right,on=['id','subject'],how='outer')
```
