```python
x= np.arange(0,4*np.pi,0.1)
y = np.sin(x)
y1=np.cos(x)
y2 = np.tan(x)
y3 = np.tanh(x)
plt.subplot(2,2,1)
plt.plot(x,y,color="g",label='Sin',linewidth=2)
plt.subplot(2,2,2)
plt.plot(x,y1,color='c',label='Cos')
plt.subplot(2,2,3)
plt.plot(x,y2,color='r',label='tan')
plt.subplot(2,2,4)
plt.plot(x,y3,color='b',label='tanh')

```
